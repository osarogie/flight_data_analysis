```python
# Months with the most flights, sorted by month number
year_map = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec",
}
df2["month_name"] = df2["month"].map(year_map)
sns.countplot(x="month_name", data=df2, order=year_map.values())
plt.xlabel("Month")
plt.ylabel("Number of Flights")
```

```python
# Merge year, month, day into one date column
df2["date"] = pd.to_datetime(df2[["year", "month", "day"]])
df2["day_of_week"] = df2["date"].dt.dayofweek.apply(
    lambda x: "Weekday" if x < 5 else "Weekend"
)
df2[["date", "day_of_week"]].tail()
```

```python
# Weekday flights vs weekend flights
sns.countplot(x="day_of_week", data=df2)
plt.xlabel("Day of Week")
plt.ylabel("Number of Flights")
```

We have more flights on


# Do you have more flights in the morning, afternoon, or evening?


```python
# Categorise flights by morning, afternoon or evening
def time_of_day(x):
    if x < 12:
        return "Morning"
    elif x < 17:
        return "Afternoon"
    else:
        return "Evening"


df2["time_of_day"] = df2["hour"].apply(time_of_day)
df2[["hour", "time_of_day"]].head()
```

```python
# Morning, afternoon, evening flights in pie chart
df2["time_of_day"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.xlabel("")
plt.ylabel("")
```

Most flights were in the morning, followed by the afternoon, and the least flights were in the evening.


```python
# busiest airports
df2["origin"].value_counts().plot(kind="bar")
plt.xlabel("Airport")
plt.ylabel("Number of Flights")
```

```python
# Which airline has the most delays?,
df2.groupby("carrier")["dep_delay"].mean().sort_values(ascending=False).plot(
    kind="bar",
)
plt.xlabel("Airline")
plt.ylabel("Average Delay (mins)")
```

```python
# Which origin has the most delays?
df2.groupby("origin")["dep_delay"].mean().sort_values(ascending=False).plot(
    kind="pie", autopct="%1.1f%%"
)
plt.xlabel("")
plt.ylabel("")
```
